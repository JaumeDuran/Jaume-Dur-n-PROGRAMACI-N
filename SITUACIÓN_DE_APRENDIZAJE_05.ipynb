```python
import sqlite3

print("\033[;34;47m"+"BIENVENIDO/A AL PROGRAMA PHYTON")
print("\033[;36;41m"+"PROGRAMA EL CUAL SE DESARROLLARÁ UNA BASE DE DATOS SOBRE PELICULAS")

nombre=input("Introduzca su nombre si es posible: ")
print("\033[;36;49m"+ "Hola "  + nombre + "," " ahora deberás elegir la opción a desear")
# Definir opciones del menú con color azul
menu = "\033[;32;49m" + "Menú:\n"

menu += "A. Añadir nuevo registro\n"
menu += "C. Consultar campo\n"
menu += "E. Eliminar registros\n"
menu += "R. Resultados\n"
menu += "S. Salir\n" 


# Mostrar el menú con formato
print(menu)

opcion = ''

while opcion not in ['A', 'C', 'E', 'R', 'S']:
    opcion = input("Elija una opción del menú (A, C, E, R, S): ").upper()

if opcion == 'A':
    print("\033[;36;48m"+"Has elegido añadir un nuevo registro.")
    # Aquí puedes añadir la lógica para añadir un nuevo registro
    # ...
    
    # Conectarse a la base de datos
    conn = sqlite3.connect('imdb_movies.db')

    # Crear un cursor
    cursor = conn.cursor()

    # Ejecutar una consulta SQL para seleccionar todos los registros de una tabla
    cursor.execute("SELECT name FROM sqlite_master WHERE type='table'")
    tablas = cursor.fetchall()

    # Obtener el nombre de la primera tabla
    nombre_tabla = tablas[0][0]

    # Ejecutar una consulta para obtener todos los registros de la tabla
    cursor.execute(f"SELECT * FROM {nombre_tabla}")
    registros = cursor.fetchall()

    # Imprimir los nombres de las columnas
    columnas = [descripcion[0] for descripcion in cursor.description]
    print(columnas)

    # Obtener los valores para el nuevo registro
    nuevo_registro = []
    for columna in columnas:
        valor = input(f"Ingrese el valor para '{columna}': ")
        nuevo_registro.append(valor)

    # Crear la consulta SQL para insertar el nuevo registro
    columnas_str = ', '.join(columnas)
    valores_str = ', '.join(['?' for _ in columnas])
    consulta = f"INSERT INTO {nombre_tabla} ({columnas_str}) VALUES ({valores_str})"

    # Ejecutar la consulta para insertar el nuevo registro
    cursor.execute(consulta, nuevo_registro)
    conn.commit()

    print("\033[;34;47m"+"Nuevo registro insertado correctamente.")

    # Cerrar la conexión y el cursor
    cursor.close()
    conn.close()


    

elif opcion == 'E':
    print("\033[;36;48m" + "Has elegido eliminar registros.")

    # Conectarse a la base de datos
    conn = sqlite3.connect('imdb_movies.db')

    # Crear un cursor
    cursor = conn.cursor()

    # Obtener el nombre de la tabla
    cursor.execute("SELECT name FROM sqlite_master WHERE type='table'")
    tablas = cursor.fetchall()
    nombre_tabla = tablas[0][0]

    # Obtener los nombres de las columnas
    cursor.execute(f"SELECT name FROM pragma_table_info('{nombre_tabla}')")
    columnas = cursor.fetchall()

    # Mostrar los nombres de las columnas disponibles
    print("Los campos disponibles son:")
    for columna in columnas:
        print(columna[0])

    # Pedir al usuario que ingrese el campo para la condición de eliminación
    campo = input("Ingrese el campo para la condición de eliminación: ")
    valor = input(f"Ingrese el valor para el campo '{campo}' a eliminar: ")

    # Crear la consulta SQL para eliminar los registros que coincidan con el campo y su valor
    consulta = f"DELETE FROM {nombre_tabla} WHERE {campo} = ?"

    # Ejecutar la consulta para eliminar los registros
    cursor.execute(consulta, (valor,))
    conn.commit()

    print("\033[;34;47m" + "Registros eliminados correctamente.")

    # Cerrar la conexión y el cursor
    cursor.close()
    conn.close()
elif opcion == 'R':
    print("\033[;36;48m" + "Has elegido ver los resultados.")

    # Conectarse a la base de datos
    conn = sqlite3.connect('imdb_movies.db')

    # Crear un cursor
    cursor = conn.cursor()

    # Obtener el nombre de la tabla
    cursor.execute("SELECT name FROM sqlite_master WHERE type='table'")
    tablas = cursor.fetchall()
    nombre_tabla = tablas[0][0]

    # Crear la consulta SQL para obtener las películas japonesas con puntuación mayor a 80
    consulta = f"SELECT * FROM {nombre_tabla} WHERE country='Japan' AND score > 80"

    # Ejecutar la consulta para obtener los registros
    cursor.execute(consulta)
    registros = cursor.fetchall()

    # Mostrar los registros encontrados
    print("Películas japonesas con puntuación mayor a 80:")
    for registro in registros:
        print(registro)

    # Cerrar la conexión y el cursor
    cursor.close()
    conn.close()


elif opcion == 'S':
 print("\033[;36;48m"+"Has elegido salir. ¡Hasta luego!")

```
